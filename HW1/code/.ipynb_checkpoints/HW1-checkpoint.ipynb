#Missing Value

```python
def countFullEntry(data):
    count = 0
    rest = []
    for each in data:
        if '' not in each:
            count += 1
            rest.append(each)
    return count, rest
```

```python
count, rest = countFullEntry(sample)
```

```python
def validCol(data):
    valid = []
    tran = data.T
    for each in tran:
        empty = len([i for i in each if i == ''])
        valid.append(1 - empty/float(len(each)))
    return valid
```

```python
valid = validCol(sample)
print valid
plt.title('non-empty percentage of each column')
plt.scatter(np.arange(len(valid)),valid)
plt.plot(np.arange(len(valid)), [0.95]*len(valid), color='red')
```

#Filling Missing Value
  1) fill in median or numerical values

```python
def fillNumeric(numeric):
    cols = numeric.T
    for col in cols:
        median = np.median(np.asarray(col[np.where(col != '')], dtype=int), axis=0)
        col[np.where(col == '')] = int(median)
    return cols.T


def fillCategoric(categoric):
    cols = categoric.T
    for col in cols:
        col[np.where(col == '')] = 'NA'
    return cols.T


def fillMissingValue(data):
    numeric = np.asarray([row[:13] for row in data])
    categoric = np.asarray([row[13:] for row in data])
    numeric = fillNumeric(numeric)
    categoric = fillCategoric(categoric)
    
    return np.column_stack((numeric,categoric))
```

```python
new_sample = fillMissingValue(sample)
```
